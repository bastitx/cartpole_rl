```python
import sympy
from sympy.parsing.latex import parse_latex
from sympy.parsing.sympy_parser import parse_expr
```

```python
xacc = parse_expr('Tm/r+mp*l*(thetadot**2*sin(theta)-thetaacc*cos(theta))-muc*Nc*sign(Nc*xdot)/(r**2*(J/r**2+mc+mp))')
```

```python
print(xacc)
```

```python
F = parse_expr('Tm/r - J*xacc/r**2')
```

```python
print(F)
```

```python
F = F.subs('xacc', xacc)
```

```python
print(F)
```

```python
thetaacc = parse_expr('(g * sin(theta) + cos(theta) * ((- F -mp*l*thetadot**2*(sin(theta)+muc*sign(Nc*xdot)*cos(theta)))/(mc+mp)+muc*g*sign(Nc*xdot)) - (mup*thetadot)/(mp*l))/(l*(4/3-(mp*cos(theta))/(mc+mp)*(cos(theta)-muc*sign(Nc*xdot))))')
```

```python
print(thetaacc)
```

```python
thetaacc = thetaacc.subs('F', F)
```

```python
print(thetaacc)
```
